```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
rc('font', family='Comic Sans MS')
d = 2.5e-6
I_0 = 1
r0 = 1
N = 3
Lambda = 5e-7
k = 2*np.pi/Lambda
theta = np.linspace(-np.pi/3,np.pi/3,10000)
def r(theta):
    return r0/np.cos(theta)
```

```python
def I(theta):
    fi = k*d*np.sin(theta)
    return I_0/r(theta)**2 *(np.sin(N*fi/2)/np.sin(fi/2))**2
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
ax1.plot(theta,I(theta),'g')
ax1.set_title("PF2 lato 2017, grupa R-3, praca domowa #4, temat C\n"
              'Paulina Marikin, nr indeksu 284494')
ax1.set_ylabel(r'I [$\frac{W}{m^2}$]')
ax1.set_xlabel(r'$\theta[rad]$')
ax1.grid()
theta_full = np.linspace(0, 2*np.pi, 10000)
ax2 = fig.add_subplot(212, projection='polar')
ax2.plot(theta_full, I(theta_full))
ax2.set_title('Porządny wykres natężenia we współrzędnych biegunowych\n')
fig.tight_layout()
fig.savefig("pd.png", dpi=800, papertype='a4')
```
