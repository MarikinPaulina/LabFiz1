```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
rc('font', family='Comic Sans MS')
d = 2.5e-6
I_0 = 1
r0 = 1
N = 3
Lambda = 5e-7
k = 2*np.pi/Lambda
theta = np.linspace(-np.pi/3,np.pi/3,10000)
def r(theta):
    return r0/np.cos(theta)
```

```python
def I(theta):
    fi = k*d*np.sin(theta)
    return I_0/r(theta)**2 *(np.sin(N*fi/2)/np.sin(fi/2))**2
```

```python
plt.plot(theta,I(theta),'g')
plt.title('Paulina Marikin praca domowa 4')
plt.ylabel(r'I$\frac{W}{m^2}$')
plt.xlabel(r'$\theta[rad]$')
plt.grid()
```

```python
theta_full = np.linspace(0, 2*np.pi, 10000)
plt.polar(theta_full, I(theta_full))
plt.title('Wykres natezenia we wspolrzednych biegunowych')
```
